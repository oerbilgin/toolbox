```python
x = np.linspace(1, 30, 100)
y = 3*x + 4
x = np.random.normal(x, scale=5)
y = np.random.normal(y, scale=5)

shape, within, without = multivar_ellipse(x,y, sigma=2)

fig, ax = plt.subplots()
# do this after you're done using the ellipse, otherwise will behave very strangely
ax.add_artist(shape) 
ax.scatter(within[0], within[1], c='lightskyblue', label='Inside ellipse')
if len(without) != 0:
    ax.scatter(without[0], without[1], c='coral', label='Outside ellipse')
ax.legend(loc='upper left', bbox_to_anchor=(1., 1.), scatterpoints=1)
plt.show()
```
