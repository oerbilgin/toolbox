```python
#make the dataset
x = np.linspace(1, 15, 50)
y = 3*x + 4
x = np.random.normal(x, scale=4)
y = np.random.normal(y, scale=4)
fig, ax = plt.subplots()
# plt.scatter(x, y, c='', edgecolors='k')

xfit, trend_fn, conf_dy, sig_dy, within, without = trendline(x,y, confint=True, sigline=True)
# plot trendline
ax.plot(xfit, trend_fn(xfit), color='k', label='%s' %(trend_fn)) 

# Plot confidence intervals of fit
ax.plot(xfit, trend_fn(xfit)+conf_dy, color='k', linestyle='--', label='95% CI') 
ax.plot(xfit, trend_fn(xfit)-conf_dy, color='k', linestyle='--') 

# Plot lines that encompass ~ 95% of data
ax.plot(xfit, trend_fn(xfit)-sig_dy, color='r', linestyle='--', label='2-sigma Data')
ax.plot(xfit, trend_fn(xfit)+sig_dy, color='r', linestyle='--') 

# plot data points that fall within or out of the range
ax.plot(within[0], within[1], color='lightskyblue', marker='o', linestyle='', label='Data within 2-sigma') # data points within the range
ax.plot(without[0], without[1], color='coral', marker='o', linestyle='', label='Data outside 2-sigma') # data points out of the range
ax.legend(loc='upper left', bbox_to_anchor=(1.05, 1.))
plt.show()
```
